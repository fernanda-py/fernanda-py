## Ejercicio 2: $\Delta \chi ^2$

```python
#read the file
X,Y = np.loadtxt(r"/content/Points_chi2_example_cov.txt").T
cov_mat = np.loadtxt(r"/content/Covariance_chi2_example_cov.txt").T
```

```python
print(X,Y)
```

```python
plt.plot(X,Y,'o',label='Data points',color='skyblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

#pcolormesh de matriz de covarianza
plt.pcolormesh(cov_mat,cmap='viridis')
plt.colorbar()
plt.title('Matriz de covarianza')
plt.show()

```

```python
#error bars
err_arr = np.sqrt(np.diag(cov_mat))
plt.figure(figsize=(8,6))
plt.errorbar(X,Y,yerr=err_arr,fmt='o',label='Datapoints',capsize=5,ecolor='skyblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
type(err_arr)
```

```python
#Prepare to fit

A_arr = np.arange(-6,4,0.1)
B_arr = np.arange(-6,4,0.1)

chi2_arr = np.zeros((len(A_arr),len(B_arr)))

for i in range(len(A_arr)):
    for j in range(len(B_arr)):
        V = A_arr[i]*X + B_arr[j] - Y
        chi2_arr[i,j] = np.dot(V.T,np.dot(np.linalg.inv(cov_mat),V))/2  #producto punto de (V^T Cov^-1 V)/2

```

```python
#bestfit
A_best,B_best = np.unravel_index(np.argmin(chi2_arr),chi2_arr.shape)
print('Best fit:',A_arr[A_best],B_arr[B_best])
```

```python
#plot data points con bestfit

plt.errorbar(X,Y,yerr=err_arr,fmt='o',label='Data',capsize=5,ecolor='skyblue')
plt.plot(X,A_arr[A_best]*X + B_arr[B_best],label='Best fit',color='red',linestyle='--')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Mejor ajuste de los datos')
plt.legend()
plt.show()
```

```python
#Contornos de confianza

Delta_chi2 = chi2_arr - np.min(chi2_arr)

plt.figure(figsize=(8,6))
plt.pcolormesh(A_arr,B_arr,Delta_chi2.T,cmap='gnuplot')
plt.plot(A_arr[A_best],B_arr[B_best],'*',label='Best fit',color='yellow',markersize = 10)
plt.colorbar()
plt.xlabel('A')
plt.ylabel('B')
plt.title(r'$\Delta \chi^2$')
plt.legend()
plt.show()
```

```python
#contornos de confianza a 1,2 y 3 sigmas

confianza_68 = 2.30
confianza_95 = 5.99
confianza_99 = 9.21

levels = [confianza_68,confianza_95,confianza_99]

A_real,B_real = 2.7,-2.3

plt.figure(figsize=(8,6))
contour = plt.contour(A_arr,B_arr,Delta_chi2.T,levels)
plt.clabel(contour, fmt={confianza_68: '$1\sigma$', confianza_95: '$2\sigma$', confianza_99: '$3\sigma$'})
plt.plot(A_arr[A_best],B_arr[B_best],'o',label='Best fit',color='blue',markersize = 5)
plt.plot(A_real,B_real,'o',label='Real',color='red',markersize = 5)
plt.xlabel('A')
plt.ylabel('B')
plt.title(r'$\Delta \chi^2$')
plt.legend()
plt.plot()
```

```python
import scipy.stats as stats
```

```python
#simulamos tiempo entre llamadas
np.random.seed(42)
true_lambda = 0.5 #llamadas cada 2 min
data = np.random.exponential(scale=1/true_lambda, size=1000) # Generamos 1000 datos de una distribuci√≥n exponencial

#histograma de los datos
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label='Datos simulados')


x= np.linspace(0,max(data),1000)
plt.plot(x,true_lambda * np.exp(-true_lambda * x), color='red', label='PDF real', linewidth=2)

print(np.mean(data))
```

```python
#usamos m√©todo de m√°xima verosimilitud para estimar lambda
def likelihood(lambda_est,data):
    L = lamda_est**len(data) * np.exp(-lambda_est * np.sum(data))
    return L

#estimamos lambda
def log_likelihood(lambda_est,data):
    return len(data) * np.log(lambda_est) - lambda_est * np.sum(data)

#estimamos lambda
lambda_est = np.linspace(0.01, 2, 100)
log_likelihood_arr = [log_likelihood(l, data) for l in lambda_est] #se itera en cada valor posible de lambda para saber cu√°l maximiza
plt.plot(lambda_est, log_likelihood_arr, label='Log-Likelihood', color='orange')
plt.axvline(true_lambda, color='green', linestyle='--', label='Lambda real')
plt.axvline(lambda_est[np.argmax(log_likelihood_arr)], color='blue', linestyle='--', label='Lambda estimado')
plt.xlabel('Lambda')
plt.ylabel('Log-Likelihood')
plt.title('Estimaci√≥n de Lambda usando M√°xima Verosimilitud')
plt.legend()

print('Lambda real:', true_lambda)
print('Lambda estimado:', lambda_est[np.argmax(log_likelihood_arr)])
plt.show()
```

La estimaci√≥n coincide con la f√≥rmula de
$\lambda _{MLE} = \frac{1}{\bar{x}}$
que nos da un valor de 0.5 = 1/2

```python
#prueba de hip√≥tesis
Lambda = 2*(log_likelihood(lambda_est[np.argmax(log_likelihood_arr)], data) - log_likelihood(true_lambda, data))
print('Valor de la prueba de hip√≥tesis:', Lambda)

p = stats.chi2.sf(Lambda, df=1)  # Usamos la funci√≥n de supervivencia para obtener el p-valor
print('P-valor:', p)
```

# Gaussian Process Regression
Ahora comenzamos con un ejercicio de GPs usando los puntos de Points_chi2_example

Ploteamos los datapoints

```python
#error bars
err_arr = np.sqrt(np.diag(cov_mat))
plt.figure(figsize=(8,6))
plt.errorbar(X,Y,yerr=err_arr,fmt='o',label='Datapoints',capsize=5,ecolor='skyblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

Vamos a generar los puntos en x sobre los que vamos a predecir

```python
x_prime = np.arange(-0.1,10.1,0.2)
print(x_prime)
```

Como solo es un ejercicio vamos a usar el kernel exponencial cuadrado

```python
def kernel(x1,x2,l = 1,alpha=30):
    x_dist = x1-x2
    return alpha**2 * np.exp(-x_dist**2/(2*l**2))
```

```python
#Construimos la matriz covarianza
K = np.zeros((len(X),len(X)))
for i in range(len(X)):
    for j in range(len(X)):
        K[i,j] = kernel(X[i],X[j])

# Agregar ruido (varianza sigma_n^2 en la diagonal)
sigma_n = 0.5  # ajusta este valor
K += (sigma_n**2) * np.eye(len(X))

#Construimos la matriz cov cruzada (t_* y t)

K1 = np.zeros((len(X),len(x_prime)))
for i in range(len(X)):
    for j in range(len(x_prime)):
        K1[i,j] = kernel(X[i],x_prime[j])

#Construimos la matriz cov de prueba (t_i*,t_j*)

K2 = np.zeros((len(x_prime),len(x_prime)))
for i in range(len(x_prime)):
    for j in range(len(x_prime)):
        K2[i,j] = kernel(x_prime[i],x_prime[j])
```

```python
#Ahora generamos la media predictiva
jitter = 1e-8 * np.eye(len(X))
K_inv = np.linalg.inv(K + jitter)
f_star = np.dot(np.dot(K1.T,K_inv),Y)

#Cov predictiva
C_star = K2 - np.dot(np.dot(K1.T,K_inv),K1)
```

```python
#Ahora vamos a la visualizaci√≥n

# --- Visualizaci√≥n ---
plt.figure(figsize=(8,5))
plt.plot(x_prime, f_star, 'pink', label='Media predictiva')
plt.fill_between(x_prime, f_star - 1.96*np.sqrt(np.diag(C_star)), f_star + 1.96*np.sqrt(np.diag(C_star)), color='lightpink',
                 alpha=0.3, label='95% de confianza')
plt.fill_between(x_prime, f_star - 2.58*np.sqrt(np.diag(C_star)), f_star + 2.58*np.sqrt(np.diag(C_star)), color='pink',
                 alpha=0.25, label='99% de confianza')
plt.plot(X, Y, 'ko', label='Datos conocidos')
plt.title('Interpolaci√≥n con Proceso Gaussiano')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, DotProduct

kernel = DotProduct(sigma_0=0.1) + WhiteKernel(noise_level=0.1)
```

```python
gaussian_process = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
gaussian_process.fit(X.reshape(-1, 1), Y)
gaussian_process.kernel_
```

```python
mean_prediction, std_prediction = gaussian_process.predict(X.reshape(-1, 1), return_std=True)

plt.plot(X, Y, 'ko', label='Datos conocidos')
plt.plot(X, mean_prediction, label="Mean prediction")
plt.fill_between(
    X.ravel(),
    mean_prediction - 1.96 * std_prediction,
    mean_prediction + 1.96 * std_prediction,
    alpha=0.5,
    label=r"95% confidence interval",
)
plt.fill_between(
    X.ravel(),
    mean_prediction - 2.58 * std_prediction,
    mean_prediction + 2.58 * std_prediction,
    alpha=0.25,
    label=r"99% confidence interval",
)
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
_ = plt.title("Gaussian process regression on noise-free dataset")
```
