## Ejercicio 2. Realización de una PDF

```python
#Tenemos que transponer el archivo, pues viene en dos columnas y para separar las dos columnas en dos vectores hay que transponer ya que python lee por línea

data = np.loadtxt(r'C:\Users\maryf\Servicio Social Cosmología Observacional\PDF_double_gauss.txt')
x,p_x = data.T #transpuesta del archivo para separar los vectores
```

```python
#diferencial de x
dx = x[1]-x[0]
print('dx:',dx)
```

```python
# revisar si pdf esta normalizada
s = np.sum(p_x*dx)
print('Suma de p(x)*dx:',s)
#se acerca a 1 bastante 
```

```python
#commulative distribution function
cumul_arr = np.zeros(len(p_x))
for i in range(len(p_x)):
    cumul_arr[i] = np.sum(p_x[:i+1]*dx)
cumul_arr /= cumul_arr[-1]
```

```python
#plotear la pdf y la cdf
plt.plot(x,p_x,label='pdf',color='skyblue')
plt.title('Probabiity Distribution Function')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.show()
```

```python
plt.plot(x,cumul_arr,label='cdf',color='lightgreen')
plt.title('Cumulative Distribution Function')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.show()
```

```python
uniform_arr = np.random.random(1000)

myrand_arr = np.zeros(1000)

for i in range(1000):
    indice = np.argmin(np.abs(cumul_arr - uniform_arr[i]))
    myrand_arr[i] = x[indice]

print(myrand_arr[:10])
```

```python
plt.hist(uniform_arr,label='Random numbers',color='lightcoral')
plt.show()
```

```python
plt.hist(myrand_arr, bins=50, density=True,range=(0,20), alpha=0.3, label='Myrand',color = 'orange' )
plt.plot(x, p_x, linestyle='--', color='red', label='PDF')
plt.xlabel('x', fontsize=15)
plt.ylabel('p(x)', fontsize=15)
plt.title('Random array vs PDF', fontsize=15)
plt.legend()
plt.show()
```

## Ejercicio 2: $\Delta \chi ^2$

```python
#read the file 
X,Y = np.loadtxt(r"C:\Users\maryf\Servicio Social Cosmología Observacional\Points_chi2_example_cov.txt").T
cov_mat = np.loadtxt(r"C:\Users\maryf\Servicio Social Cosmología Observacional\Covariance_chi2_example_cov.txt").T
```

```python
print(X,Y)
```

```python
plt.plot(X,Y,'o',label='Data points',color='skyblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

#pcolormesh de matriz de covarianza 
plt.pcolormesh(cov_mat,cmap='viridis')
plt.colorbar()
plt.title('Matriz de covarianza')
plt.show()

```

```python
#error bars
err_arr = np.sqrt(np.diag(cov_mat))
plt.figure(figsize=(8,6))
plt.errorbar(X,Y,yerr=err_arr,fmt='o',label='Datapoints',capsize=5,ecolor='skyblue')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
#Prepare to fit

A_arr = np.arange(-6,4,0.1)
B_arr = np.arange(-6,4,0.1)

chi2_arr = np.zeros((len(A_arr),len(B_arr)))

for i in range(len(A_arr)):
    for j in range(len(B_arr)):
        V = A_arr[i]*X + B_arr[j] - Y
        chi2_arr[i,j] = np.dot(V.T,np.dot(np.linalg.inv(cov_mat),V))/2  #producto punto de (V^T Cov^-1 V)/2

```

```python
#bestfit
A_best,B_best = np.unravel_index(np.argmin(chi2_arr),chi2_arr.shape)
print('Best fit:',A_arr[A_best],B_arr[B_best])
```

```python
#plot data points con bestfit

plt.errorbar(X,Y,yerr=err_arr,fmt='o',label='Data',capsize=5,ecolor='skyblue')
plt.plot(X,A_arr[A_best]*X + B_arr[B_best],label='Best fit',color='red',linestyle='--')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Mejor ajuste de los datos')
plt.legend()
plt.show()
```

```python
#Contornos de confianza

Delta_chi2 = chi2_arr - np.min(chi2_arr)

plt.figure(figsize=(8,6))
plt.pcolormesh(A_arr,B_arr,Delta_chi2.T,cmap='gnuplot')
plt.plot(A_arr[A_best],B_arr[B_best],'*',label='Best fit',color='yellow',markersize = 10)
plt.colorbar()
plt.xlabel('A')
plt.ylabel('B')
plt.title(r'$\Delta \chi^2$')
plt.legend()
plt.show()
```

```python
#contornos de confianza a 1,2 y 3 sigmas

confianza_68 = 2.30
confianza_95 = 5.99
confianza_99 = 9.21

levels = [confianza_68,confianza_95,confianza_99]

A_real,B_real = 2.7,-2.3

plt.figure(figsize=(8,6))
contour = plt.contour(A_arr,B_arr,Delta_chi2.T,levels)
plt.clabel(contour, fmt={confianza_68: '$1\sigma$', confianza_95: '$2\sigma$', confianza_99: '$3\sigma$'})
plt.plot(A_arr[A_best],B_arr[B_best],'o',label='Best fit',color='blue',markersize = 5)
plt.plot(A_real,B_real,'o',label='Real',color='red',markersize = 5)
plt.xlabel('A')
plt.ylabel('B')
plt.title(r'$\Delta \chi^2$')
plt.legend()
plt.plot()
```

```python
import scipy.stats as stats
```

```python
#simulamos tiempo entre llamadas
np.random.seed(42)
true_lambda = 0.5 #llamadas cada 2 min
data = np.random.exponential(scale=1/true_lambda, size=1000) # Generamos 1000 datos de una distribución exponencial

#histograma de los datos
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label='Datos simulados')


x= np.linspace(0,max(data),1000)
plt.plot(x,true_lambda * np.exp(-true_lambda * x), color='red', label='PDF real', linewidth=2)

print(np.mean(data))
```

```python
#usamos método de máxima verosimilitud para estimar lambda
def likelihood(lambda_est,data):
    L = lamda_est**len(data) * np.exp(-lambda_est * np.sum(data))
    return L

#estimamos lambda
def log_likelihood(lambda_est,data):
    return len(data) * np.log(lambda_est) - lambda_est * np.sum(data)

#estimamos lambda
lambda_est = np.linspace(0.01, 2, 100)
log_likelihood_arr = [log_likelihood(l, data) for l in lambda_est] #se itera en cada valor posible de lambda para saber cuál maximiza
plt.plot(lambda_est, log_likelihood_arr, label='Log-Likelihood', color='orange')
plt.axvline(true_lambda, color='green', linestyle='--', label='Lambda real')
plt.axvline(lambda_est[np.argmax(log_likelihood_arr)], color='blue', linestyle='--', label='Lambda estimado')
plt.xlabel('Lambda')
plt.ylabel('Log-Likelihood')
plt.title('Estimación de Lambda usando Máxima Verosimilitud')
plt.legend()

print('Lambda real:', true_lambda)
print('Lambda estimado:', lambda_est[np.argmax(log_likelihood_arr)])
plt.show()
```

La estimación coincide con la fórmula de 
$\lambda _{MLE} = \frac{1}{\bar{x}}$
que nos da un valor de 0.5 = 1/2

```python
#prueba de hipótesis
Lambda = 2*(log_likelihood(lambda_est[np.argmax(log_likelihood_arr)], data) - log_likelihood(true_lambda, data))
print('Valor de la prueba de hipótesis:', L)

p = stats.chi2.sf(Lambda, df=1)  # Usamos la función de supervivencia para obtener el p-valor
print('P-valor:', p)
```
